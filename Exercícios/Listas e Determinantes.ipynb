1. (5.0) Escreva uma função em Python para solucionar o seguinte problema: dadas 2
listas com n números entre 0 e 9, interpretadas como dois números inteiros de n
dígitos, calcular uma lista que representa a somatória dos dois números.

```python
a = [ 2, 3, 4]
b = [9, 9, 0, 0]
def soma_de_listas(primeira_lista, segunda_lista):
   numero1 = ''
   numero2 = ''
   somatoria = []
   for i in primeira_lista:
     numero1 += str(i)
   for i in segunda_lista:
     numero2 += str(i)
   soma = int(numero1) + int(numero2)
   for i in str(soma):
     somatoria.append(int(i))
   print(somatoria)
soma_de_listas(a, b)
```

2. (5.0) Faça uma função em Python que verifique a seguinte propriedade do
determinante de uma matriz fornecida pelo usuário.
“Se todos os elementos de uma linha ou coluna são iguais à zero, o determinante da
matriz é nulo.”
A função deve apresentar uma mensagem informando se o determinante da matriz
fornecida é ou não igual a zero segundo a propriedade acima.

```python
matriz = [[1, 4, 5], [0, 0, 0], [0, 2, 6]]

def determinante(matriz):
  soma_colunas = [0]*len(matriz[0])
  for i, elemento in enumerate(matriz):
    if sum(matriz[i]) == 0:
      return'A determinante é nula'
    for j, n in enumerate(elemento):
        soma_colunas[j] += n
  for i in soma_colunas:
    if i == 0:
      return'A determinante é nula'
  return 'A determinante NÃO é nula'

print(determinante(matriz))
```
