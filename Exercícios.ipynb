1. (4.0) Escrever um programa em Python que lê o público total de um jogo de futebol e fornecer a renda do jogo, sabendo-se que havia 4 tipos de ingressos assim distribuídos: popular 10% a R$10,00, geral 50% a RS50,00, arquibancada 30% a RS100,00 e cadeiras 10% a RS 200,00.

```python
publitotal = int(input('Digite o público total: '))
i10 = (publitotal * 10/100)
i50 = (publitotal * 50/100)
i100 = (publitotal * 30/100)
i200 = (publitotal * 10/100)
renda = (i10*10) + (i50*50) + (i100*100) + (i200*200)
print(f'A renda total é de R${renda}')
```

2. (6.0) Faça um programa que leia os valores de 30 medidas de massas (no formato de toneladas, quilos e gramas) e ao final mostre a soma e a média das medidas informadas. A seguir é apresentado um exemplo para ilustrar o entendimento do problema. Considere t como toneladas, k como quilos e g como gramas.

OBS1: a entrada dos dados deve ser feita para cada unidade, ou seja, ler primeiro o valor das toneladas, depois o valor dos quilos e, por fim, o valor das gramas para cada medida de massa.

OBS2: não utilizar funções ou listas no desenvolvimento da solução.

OBS3: após a realização das operações, caso haja valores que possam ser agrupados fazer as devidas transformações conforme o anexo.




```python
totalton = totalkg = totalg = 0
for c in range(0, 2):
  ton = int(input('Digite as toneladas: '))
  kg =  int(input('Digite os quilogramas: '))
  g =  int(input('Digite as gramas: '))
  if kg >= 1000:
    ton += kg//1000
    kg = kg%1000
  if g >= 1000:
    kg += g//1000
    g = g%1000
  totalton += ton
  totalkg += kg
  totalg += g
mediaton = totalton/2
mediakg = totalkg/2
mediag = totalkg/2
print('='*10, 'SOMA', '='*10)
print(f'{totalton}T  {totalkg}kg {totalg}g' )
print('='*10, 'MÉDIA', '='*10)
print(f'{mediaton}T  {mediakg}kg {mediag}g' )
```
